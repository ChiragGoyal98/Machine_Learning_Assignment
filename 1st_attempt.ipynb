```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
from matplotlib import pyplot as plt
#import seaborn as sns
```

```python
ds=pd.read_csv("musk_csv.csv", delimiter=",")
ds.describe()
```

```python
ds.head()
```

```python
ds
```

# Randomizing the dataset 


```python
ds_1 = ds.sample(frac=1).reset_index(drop=True)
ds_1.head(20)
```

# Dropping irrelevant columns

```python
ds_1.drop(ds_1.columns[[0,1,2]], axis = 1, inplace = True) 
ds_1
```

# One Hot encoding

```python
Y = ds_1.iloc[:,166]
Y=Y.to_numpy()
print(Y.shape)
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
Y = ohe.fit_transform(Y.reshape(-1,1)).toarray()
Y
```

```python
X= ds_1.drop(ds_1.columns[166], axis = 1)
X.shape, Y.shape
```

```python
X.head(20)
```

# Split the data into train and test (80:20)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

# Scaling the remaining 166 columns into a rane of -1 to 1 
## using MIn Max Scaler

```python
#scaling the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range =(-1,1))
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
#Logistic Regression
y_train = []
for i in range(len(Y_train)):
    y_train.append(np.argmax(Y_train[i]))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(C=2)
clf.fit(X_train, y_train)
```

```python
y_test = []
for i in range(len(Y_test)):
    y_test.append(np.argmax(Y_test[i]))
y_test
```

```python
ypred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ypred, y_test))
```

# Using a Neural Network
## since LR didnt give better results.

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
# Neural network
model = Sequential()
model.add(Dense(16, input_dim=166, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(64, activation='relu'))
#model.add(Dense(64, activation='sigmoid'))
model.add(Dense(2, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.2)
```

```python
model.summary()
```

```python
history = model.fit(X_train, Y_train, epochs=100, batch_size=64, validation_data = (X_val,Y_val))
```

```python
y_pred = model.predict(X_test)
pred = list()
for i in range(len(y_pred)):
    pred.append(np.argmax(y_pred[i]))
pred = np.asarray(pred)
Y_pred = ohe.fit_transform(pred.reshape(-1,1)).toarray()
Y_pred
```

```python
model.evaluate(X_test,Y_test)
```

# ANN gave an accuracy of 99%

```python
Y_test, y_pred
```

```python
print(classification_report(Y_test, Y_pred))
```
